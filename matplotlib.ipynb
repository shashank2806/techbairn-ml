## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Line Plot

```python
# Generate 100 points between 0 and 10
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y, 'red')
plt.xlabel('x')
plt.ylabel('Sin(x)')
plt.title('Graph of Sin(x)')
plt.show()
```

```python
y = np.cos(x)
```

```python
plt.plot(x, y, 'green')
plt.xlabel('x')
plt.ylabel('Cos(x)')
plt.title('Graph of Cos(x)')
plt.xticks(np.arange(0,10, 0.5), rotation=-90)
# plt.savefig('abc.png')
plt.show()
```

### Scatter Plot

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([10,5,7,6,11,6,7])
```

```python
plt.scatter(x,y, marker='o', c='red')
# plt.scatter([3,4,5], [1,2,3])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Histograms

```python
y = [0, 10,10,80, 81, 82, 83, 84, 85, 86, 71,71,72, 60, 100]
```

```python
plt.hist(y, bins=5)
plt.ylabel('freqency')
plt.show()
```
